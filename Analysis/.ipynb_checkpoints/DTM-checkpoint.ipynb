```python
#Import wordcloud library for visualization
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
```

```python
#load the string list
import pickle

with open ('stringData.txt', 'rb') as fp:
    stringData = pickle.load(fp)
```

```python
#Set wordcloud dimensions
wordCloud = WordCloud(width = 800,
                      height = 800,
                      background_color = 'white',
                      stopwords = set(STOPWORDS),
                      colormap = 'Dark2', 
                      max_font_size = 120,
                      random_state = 21)

wordCloud.generate(stringData)
```

```python
#set the dimensions

plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordCloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
  
plt.show()
```
