# What makes Christmas merrier?

```python
from bs4 import BeautifulSoup #Library for web scraping
import requests #url request for webscraping
import pickle #creation and loading of files
import re #String manipulation for later cleaning
import string #String manipulation library
import pandas as pd #Library for dataframe creation
import matplotlib.pyplot as plt #Data Visualizations
```

```python
#List down references
url = [
    "https://www.beliefnet.com/faiths/religious-observances/christmas/6-things-parents-can-do-to-make-christmas-merrier.aspx",
    "https://us.woodgeekstore.com/blogs/news/christmas-gifts",
    "https://minibusinessnews.com/how-to-make-a-christmas-even-merrier/",
    "https://www.realbuzz.com/articles-interests/festive-health-fitness/article/10-ways-to-make-this-your-best-christmas-ever/",
    "https://www.insider.com/things-you-should-always-do-on-christmas-2017-12#spend-time-with-loved-ones-3",
    "https://www.lifeandexperiences.com/how-to-do-good-this-christmas/",
    "https://www.herhappyheart.com/christmas-bucket-list/",
    "https://www.countryliving.com/entertaining/g2801/christmas-bucket-list/?slide=1"
]

author = [
    "Mark Driscoll","Woodgeek","Sunny95mishra95","RealBuzz","Madeline Diamond", "Amelie Lawrence","Klara Donovan", "Country Living Staff"
]
```

```python
#Set to transcript
def url_to_transcript(url):
    page = requests.get(url).text
    soup = BeautifulSoup(page,'html.parser')
    if url == "https://www.beliefnet.com/faiths/religious-observances/christmas/6-things-parents-can-do-to-make-christmas-merrier.aspx":
        text = [ p.text.lower() for p in soup.find(class_="beliefnet-gallery entry").find_all(class_="hide")]
    elif url == "https://us.woodgeekstore.com/blogs/news/christmas-gifts":
        text = [ p.text.lower() for p in soup.find(class_="article_content clearfix").find_all('h2')]
    elif url == "https://minibusinessnews.com/how-to-make-a-christmas-even-merrier/":
        text = [ p.text.lower() for p in soup.find(class_="entry-content").find_all('li')]
    elif url == "https://www.realbuzz.com/articles-interests/festive-health-fitness/article/10-ways-to-make-this-your-best-christmas-ever/":
        text = [ p.text.lower() for p in soup.find_all(class_='numbered-heading__table-layout-cell')]
    elif url == "https://www.insider.com/things-you-should-always-do-on-christmas-2017-12#spend-time-with-loved-ones-3":
        text = [ p.text.lower() for p in soup.find_all(class_='slide-title-text')]
    elif url == "https://www.lifeandexperiences.com/how-to-do-good-this-christmas/":
        text = [ p.text.lower()  for p in soup.find_all('h2')]
    elif url == "https://www.herhappyheart.com/christmas-bucket-list/":
        text = [ p.text.lower() for p in soup.find_all("p")[16:46]]
    elif url == "https://www.countryliving.com/entertaining/g2801/christmas-bucket-list/?slide=1":
        text = [ p.text.lower() for p in soup.find_all(class_= "slideshow-slide-hed")]

        
    return text
        
    
```

```python
transcript = [url_to_transcript(u) for u in url]
```

```python
#create directory
!mkdir transcripts
#write each text data of url to a textfile
for i, c in enumerate(author):
  with open("transcripts/" + c + ".txt", "wb") as file:
    pickle.dump(transcript[i], file)
```

```python
#load files
data = {}
for i, c in enumerate(author):
  with open("transcripts/" + c + ".txt", "rb") as file:
    data[c] = pickle.load(file)
```

```python
#set column width of dataframe
pd.set_option('max_colwidth',500)
```

```python
def data_clean(key):
    text = re.sub(r'[0-9]', '', key)
    text = re.sub(r'[.\n)(?!+-]', '', text)
    return text.strip()
```

```python
#Create DataFrame and Data Cleaning

for auth in data:
    for key in data[auth]:
        data[auth][data[auth].index(key)] = data_clean(key)

            
    df = pd.DataFrame(data[auth],columns = [auth])
```

```python
from textblob import TextBlob

#Create a function to get the subjectivity
def getSubjectivity(text):
    return TextBlob(text).sentiment.subjectivity

#Create a function to get the polarity
def getPolarity(text):
    return TextBlob(text).sentiment.polarity
```

```python
#View DataFrame
for auth in data:
    for key in data[auth]:
        df = pd.DataFrame(data[auth],columns = [auth])
        df['Subjectivity'] = df[auth].apply(getSubjectivity)
        df['Polarity'] = df[auth].apply(getPolarity)

    display(df)
    
```

```python

# Plot the polarity and subjectivity
plt.figure(figsize=(8,6))
plt.title('Sentiment Analysis')
plt.xlabel('Polarity')
plt.ylabel('Subjectivity')

for auth in data:
    for key in data[auth]:
        pass
    df = pd.DataFrame(data[auth],columns = [auth])
    df['Subjectivity'] = df[auth].apply(getSubjectivity)
    df['Polarity'] = df[auth].apply(getPolarity)

    
    for i in range(0, df.shape[0]):
        plt.scatter(df['Polarity'][i], df['Subjectivity'][i], color='Blue')
    plt.show()
```

```python
#Get analysis of the polarity
def getPolAnalysis(score):
    if score < 0:
        return 'Negative'
    elif score == 0:
        return 'Neutral'
    else:
        return 'Positive'

#Get analysis of the Subjective
def getSubAnalysis(score):
    if score < 0:
        return 'Fact'
    elif score == 0:
        return 'Neutral'
    else:
        return 'Opinion'

df['SubjectivityAnalysis'] = df['Subjectivity'].apply(getSubAnalysis)
df['PolarityAnalysis'] = df['Polarity'].apply(getPolAnalysis)
```

```python
#Create columns for the subjectivity and polarity results
for auth in data:
    for key in data[auth]:
        df = pd.DataFrame(data[auth],columns = [auth])
        df['Subjectivity'] = df[auth].apply(getSubjectivity)
        df['Polarity'] = df[auth].apply(getPolarity)
        df['SubjectivityAnalysis'] = df['Subjectivity'].apply(getSubAnalysis)
        df['PolarityAnalysis'] = df['Polarity'].apply(getPolAnalysis)
        
    display(df)
```

```python
#Analysis average
def perAnalysis(sentiment,df):
    return round((sentiment.shape[0] / df.shape[0]) * 100 , 1)
    
```

```python
#display overall average of every author's sentiment result
for auth in data:
    for key in data[auth]:
        df = pd.DataFrame(data[auth],columns = [auth])
        df['Subjectivity'] = df[auth].apply(getSubjectivity)
        df['Polarity'] = df[auth].apply(getPolarity)
        df['SubjectivityAnalysis'] = df['Subjectivity'].apply(getSubAnalysis)
        df['PolarityAnalysis'] = df['Polarity'].apply(getPolAnalysis)

    display(df)
    
    # Get the percentage of Opinion Analysis 
    subOpinion = df[df.SubjectivityAnalysis == 'Opinion']
    subOpinion = subOpinion['SubjectivityAnalysis']
    # Get the percentage of Opinion Analysis 
    subFact = df[df.SubjectivityAnalysis == 'Fact']
    subFact = subFact['SubjectivityAnalysis']
    # Get the percentage of Opinion Analysis 
    subNeutral = df[df.SubjectivityAnalysis == 'Neutral']
    subNeutral = subNeutral['SubjectivityAnalysis']

    
    # Get the percentage of Negative Analysis 
    polNegative = df[df.PolarityAnalysis == 'Negative']
    polNegative = polNegative['PolarityAnalysis']
    # Get the percentage of Opinion Analysis 
    polPositive = df[df.PolarityAnalysis == 'Positive']
    polPositive = polPositive['PolarityAnalysis']
    # Get the percentage of Opinion Analysis 
    polNeutral = df[df.PolarityAnalysis == 'Neutral']
    polNeutral = polNeutral['PolarityAnalysis']
    

    # plot and visualize the counts
    plt.title("Sentiment Analysis: " + auth)
    plt.xlabel('Sentiment')
    plt.ylabel('Counts')
    
    
    df["SubjectivityAnalysis"].value_counts().plot(kind='bar')
    plt.show()
    
    print("\nSubjectivity Analysis\n")
    print("Opinion:",perAnalysis(subOpinion,df))
    print("Fact:",perAnalysis(subFact,df)) 
    print("Neutral:",perAnalysis(subNeutral,df))
    print()
    
    df["PolarityAnalysis"].value_counts().plot(kind='bar')
    plt.show()
    
    print("\nPolarity Analysis\n")
    print("Negative:",perAnalysis(polNegative,df))
    print("Positive:",perAnalysis(polPositive,df))
    print("Neutral:",perAnalysis(polNeutral,df))
```
